# Marker genes

Let us compute a ranking for the highly differential genes in each cluster. For this, by default, the .raw attribute of AnnData is used in case it has been initialized before. The simplest and fastest method to do so is the t-test.

```python
import numpy as np
import pandas as pd
import scanpy as sc


sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
```

```python
sc.settings.set_figure_params(dpi=80)
```

Read in the clustered data object.

```python
cdata = sc.read_h5ad('./../write/scanpy/mnn_corrected_clustered_3pbmc.h5ad')
```

```python
print(cdata.shape)
print(cdata.raw.shape)
```

The clustering with resolution 0.6 seems to give a reasonable number of clusters, so we will use that clustering for all DE tests.

### T-test 

```python
sc.tl.rank_genes_groups(cdata, 'louvain_0.6', method='t-test', key_added = "t-test")
sc.pl.rank_genes_groups(cdata, n_genes=25, sharey=False, key = "t-test")
```

```python
# results are stored in the cdata.uns["t-test"] slot
cdata
```

```python
sc.settings.verbosity = 2  # reduce the verbosity
```

### Wilcoxon rank-sum

The result of a [Wilcoxon rank-sum (Mann-Whitney-U)](https://de.wikipedia.org/wiki/Wilcoxon-Mann-Whitney-Test) test is very similar. We recommend using the latter in publications, see e.g., [Sonison & Robinson (2018)](https://doi.org/10.1038/nmeth.4612). You might also consider much more powerful differential testing packages like MAST, limma, DESeq2 and, for python, the recent diffxpy.

```python
sc.tl.rank_genes_groups(cdata, 'louvain_0.6', method='wilcoxon', key_added = "wilcoxon")
sc.pl.rank_genes_groups(cdata, n_genes=25, sharey=False, key="wilcoxon")
```

### Logistic regression test

As an alternative, let us rank genes using logistic regression. For instance, this has been suggested by [Natranos et al. (2018)](https://doi.org/10.1101/258566). The essential difference is that here, we use a multi-variate appraoch whereas conventional differential tests are uni-variate. [Clark et al. (2014)](https://doi.org/10.1186/1471-2105-15-79) has more details.

```python
sc.tl.rank_genes_groups(cdata, 'louvain_0.6', method='logreg',key_added = "logreg")
sc.pl.rank_genes_groups(cdata, n_genes=25, sharey=False, key = "logreg")
```

### Visualization

There are several ways to visualize the expression of top DE genes. Here we will plot top 5 genes per cluster from Wilcoxon test as heatmap, dotplot, violin plot or matrix.

```python
sc.pl.rank_genes_groups_heatmap(cdata, n_genes=5, key="wilcoxon", groupby="louvain_0.6", show_gene_labels=True)
```

```python
sc.pl.rank_genes_groups_dotplot(cdata, n_genes=5, key="wilcoxon", groupby="louvain_0.6")
```

```python
sc.pl.rank_genes_groups_stacked_violin(cdata, n_genes=5, key="wilcoxon", groupby="louvain_0.6")
```

```python
sc.pl.rank_genes_groups_matrixplot(cdata, n_genes=5, key="wilcoxon", groupby="louvain_0.6")
```

### Compare specific clusters

We can also do pairwise comparisons of individual clusters on one vs many clusters. 

For instance, clusters 2 & 6 have very similar expression profiles.

```python
sc.tl.rank_genes_groups(cdata, 'louvain_0.6', groups=['2'], reference='6', method='wilcoxon')
sc.pl.rank_genes_groups(cdata, groups=['2'], n_genes=20)
```

Plot as violins for those two groups.

```python
sc.pl.rank_genes_groups_violin(cdata, groups='2', n_genes=10)
```

```python
# plot the same genes as violins across all the datasets.

# convert numpy.recarray to list
mynames = [x[0] for x in cdata.uns['rank_genes_groups']['names'][:10]]
sc.pl.stacked_violin(cdata, mynames, groupby = 'louvain_0.6')
```
